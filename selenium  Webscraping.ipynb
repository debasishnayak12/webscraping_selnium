```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
#Microsoft Edge
#driver = webdriver.Edge()
# Google Chrome
#driver = webdriver.Chrome()
```

```python
driver = webdriver.Chrome()
```

```python
main_url='https://www.naukri.com/'
```

```python
job='Data Scientist'
loc='Banglore'
```

```python
driver.get(main_url)
time.sleep(3)
```

```python
deg=driver.find_element(By.XPATH,"//input[@class='suggestor-input ']") #finding the  desingnation  path
deg.clear() #cear the input area
deg.send_keys(job)  #sending job name
loc_t=driver.find_element(By.XPATH,"//div[@class='locationSugg']/div/div/div/div/div/input") #Finding the loaction Path
loc_t.clear()#cear the input area
loc_t.send_keys(loc) #sending the location Name
driver.find_element(By.XPATH,"//div[@class='qsbSubmit']").click() #click on Serch Button
time.sleep(5)
```

```python
#finding number of Jobs
num_job=int(driver.find_element(By.XPATH,"//span[@class='fleft count-string mr-5 fs12']").text.split(' ')[-1])
num_job
job_list=[]
```

```python
def find_jb_link():
    jb_lnk=driver.find_elements(By.XPATH,"//div[@class='jobTupleHeader']/div/a")
    for i in jb_lnk:
        job_list.append(i.get_attribute('href'))
```

```python
while len(job_list)!=num_job:
    try:
        find_jb_link()
        driver.find_element(By.XPATH,"//div[@class='mt-36 mb-54']/div/a[2]").click()
        time.sleep(5)
    except:
        break
len(job_list)
```

```python
jb_ttl,exp,salary,location,job_decs=[],[],[],[],[]
```

```python
acc_lnk='https://www.naukri.com/job-listings-gn-strategy-mc-t-o-nlp-data-science-analyst-accenture-solutions-pvt-ltd-mumbai-hyderabad-secunderabad-pune-gurgaon-gurugram-chennai-bangalore-bengaluru-1-to-3-years-270823902270?src=jobsearchDesk&sid=16941539149904606&xp=1&px=1'
```

```python
driver.get(acc_lnk)
```

```python
def accenture_ibm():
    try:
        jb_ttl.append(driver.find_element(By.XPATH,"//div[@class='jdContainer job-desc']/header/div/h1").text)
    except:
        jb_ttl.append('null')
    try:
       #Experience
        exp.append(driver.find_element(By.XPATH,"//div[@class='jdContainer job-desc']/header/div[2]/span").text) 
    except:
        exp.append('null')
    try:
        #salary
        salary.append(driver.find_element(By.XPATH,"//div[@class='jdContainer job-desc']/header/div[3]/span").text)
    except:
        salary.apeend('null')
    try:
        temp_list=[]
        for i in driver.find_elements(By.XPATH,"//div[@class='jdContainer job-desc']/header/div[4]/div/div/span/a"):
            temp_list.append(i.text)
        location.append(temp_list)
        del temp_list
    except:
        location.append('null')
    try:
        job_decs.append(driver.find_element(By.XPATH,"//div[@class='nConfig_textblock ']/div[2]").text)
    except:
        job_decs.append('null')
    return None
```

```python
def Noukri():
    try:
        jb_ttl.append(driver.find_element(By.XPATH,"//div[@class='styles_jhc__top__BUxpc']/div/header/h1").text)
    except:
        jb_ttl.append('null')
    try:
       #Experience
        exp.append(driver.find_element(By.XPATH,"//div[@class='styles_jhc__top__BUxpc']/div[2]/div/div[1]").text) 
    except:
        exp.append('null')
    try:
        #salary
        salary.append(driver.find_element(By.XPATH,"//div[@class='styles_jhc__top__BUxpc']/div[2]/div/div[2]").text)
    except:
        salary.apeend('null')
    try:
        location.append(driver.find_element(By.XPATH,"//div[@class='styles_jhc__top__BUxpc']/div[2]/div[3]/span").text)
    except:
        location.append('null')
    try:
        job_decs.append(driver.find_element(By.XPATH,"//div[@class='styles_JDC__dang-inner-html__h0K4t']").text)
    except:
        job_decs.append('null')
    return None
```

```python
cnt=0
for i in job_list:
    driver.get(i)
    time.sleep(5)
    try:
        Noukri()
        print('Noukri')
    except:
        try:
            accenture_ibm()
            print('accenture_ibm')
        except:
            print('Other')
            jb_ttl.append('null')
            exp.append('null')
            salary.append('null')
            location.append('null')
            job_decs.append('null')
    print(cnt)
    cnt=cnt+1
```
